```python
# Tool to craw all files with a specific extension from a directory and subdirectories

# Load dependancies
from tkinter import filedialog
from tkinter import * # File dialog
import os
import matplotlib.pyplot as plt # For plotting
import numpy
import numpy as np
import cv2
from skimage.color import rgb2gray
from skimage import img_as_ubyte
import glob # For importing multiple files
```

```python
# Determine operation system
# 'posix' = MacOS
# 'nt' = WIN
operation_sys = os.name
```

```python
# Select file extension
select_ext = '*.jpg'
```

```python
# Get user-specific folder selection
root = Tk()
root.withdraw()
folder_selected = filedialog.askdirectory()
# root.destroy() # Destroy root window
```

```python
# ------------------------------------------------------------------------------------------ #
# Check for Folder

# Export folder name
export_folder_name = 'crop'
# Path for crop folder
cropFolder = os.path.join(folder_selected, export_folder_name)
# Export folder
if os.path.exists( cropFolder ):
    print('Output: ##### Export crop folder is available. #####')
else:
    os.makedirs( cropFolder )
    print('Output: ##### Create export folder called crop. #####')
# ------------------------------------------------------------------------------------------ #
```

```python
image_list = []
counter = 0
n_total = len( glob.glob(os.path.join(folder_selected, select_ext)) )
for filename in glob.glob( os.path.join(folder_selected, select_ext) ):
    counter = counter + 1
   # Load image
    img = cv2.imread(filename)
    # Obtain filename
    th_fname = os.path.basename(filename)
    # Convert to gray
    gray=rgb2gray(img)
    # Convert to uint8
    gray = img_as_ubyte(gray)
    #plt.figure(figsize=(18, 16))
    #plt.imshow(gray, cmap = 'gray')
    # Size of the image in pixels (size of orginal image)
    y, x = gray.shape
    # Define the pixel to truncate in height
    h = 10
    # Define the pixel to truncate in width
    w = 10
    # Crop top and right side
    crop_img = gray[h:y, 0:x-w]
    # plt.figure(figsize=(18, 16))
    # plt.imshow(crop_img, cmap ='gray')
    # Save cropped image in crop folder
    cv2.imwrite(os.path.join(cropFolder, th_fname), crop_img)
    # Print output
    print('Output: ##### cropped and saved: %d out of %d images. #####' % (counter, n_total) ) 
    
print('Output: ##### Cropping complete. #####')
```
